## las variables
1. CP -- costo del producto
2. CD -- cantidad de dinero
3. Cambio

## formulas
$Cambio = CD - CP$


## diagrama de flujo lucidchart
<img src="https://github.com/nahumflores128/-calcular-cambio-con-if-else/blob/main/DF%20if%20else%20cambio.png?raw=true" width="400" height="500" alt="Gatito de ejemplo">


```python
# para cuando es verdadero (si)
# Calculando el cambio usando if else
# datos de entrada CD, CP
# mostrar cambio si es falso
# si es verdadeero mostrar No alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD < CP:
    print("no alcanza")
    print("fin del progama")
    print("progama realizado por Nahum flores NL = 18")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
```

```python
# para cuando es falso (no)
# Calculando el cambio usando if else
# datos de entrada CD, CP
# mostrar cambio si es falso
# si es verdadeero mostrar No alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD < CP:
    print("no alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
    print("fin del progama")
    print("progama realizado por Nahum flores NL = 18")
```
